```python
import os
from pathlib import Path
from pytube import YouTube
```

```python
url = 'https://www.youtube.com/watch?v=rq3bZzLhyOM'
yt = YouTube(url)
video_path = Path('videos')
if not os.path.exists: os.makedirs(video_path)
# progressive - True (delete)
yt.streams.filter(progressive=True, file_extension='mp4').order_by(
        'resolution').desc().first().download(output_path=video_path)
```

```python
import cv2
from PIL import Image

i = 0
cap = cv2.VideoCapture((video_path / 'The Second Waltz (Full Version) - Guitar Lesson + TAB.mp4').__str__())   # /dev/video0
while i<500:
  ret, frame = cap.read()
  if not ret:
    break
  pil_img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
  i+=1
pil_img
```

```python
path_images = Path('images')
if not os.path.exists(path_images): os.makedirs(path_images)
vidcap = cv2.VideoCapture(str(video_path / 'The Second Waltz (Full Version) - Guitar Lesson + TAB.mp4'))
success,image = vidcap.read()
count = 0
while success:
    cv2.imwrite(str(path_images / ("frame%d.jpg" % count)), image)     # save frame as JPEG file      
    success, image = vidcap.read()
    # print('Read a new frame: ', success)
    count += 1
```
